**Análise para obter um recorte de tempo**

Filtramos um intervalo específico de rideids e fizemos alguns joins pois gostariamos de saber se schedule, create e update eram feitos "ao mesmo tempo" ou se elas possuiam um intervalo (entre origem e destino) e apartir desse recorte termos um possível tempo de viagem (nossa hipóese estava errada, são feitas ao mesmo tempo, na verdade nem teria sentido serem feitas posteriormente mas não custava conferir

```python
# Filtrar os rideid no intervalo de 1685719 a 1685755
df_filtrado = df_rideaddress[(df_rideaddress['rideid'] >= 1685719) & (df_rideaddress['rideid'] <= 1685755)]

# Exibir o resultado
display(df_filtrado)
```

```python
# Fazer o join com a tabela df_ride para mais contexto
df_joinado = pd.merge(df_filtrado, df_ride, on='rideid', how='left')

# Exibir o DataFrame
display(df_joinado)
```

```python
# Filtro do intervalo de rideid na rideaddress
df_filtrado = df_rideaddress[(df_rideaddress['rideid'] >= 1685719) & (df_rideaddress['rideid'] <= 1685755)]

# Join com df_ride para trazer a coluna schedule
df_joinado = pd.merge(df_filtrado, df_ride[['rideid', 'schedule', 'create', 'updated']], on='rideid', how='left')

# Exibir o DataFrame
display(df_joinado)
```

**Seleção dos dados a serem utilizados**

A base product foi desconsiderada, pois a mesma se apresentava bem inconsistente a as colunas que decedimos usar "providerid" e "description" não eram de fato tão relevantes quanto achavamos

```python

# Ride: mantido apenas RideID, Schedule, Create, Updated
df_ride = df_ride[['rideid', 'schedule', 'create', 'updated']]

# RideAddress: mantido apenas Number, Lat, Lng, RideAddressTypeID, RideID
df_rideaddress = df_rideaddress[['number', 'lat', 'lng', 'rideaddresstypeid', 'rideid']]

# RideEstimative: mantido RideID, ProductID e Price
df_rideestimative = df_rideestimative[['rideid', 'productid', 'price']]
```

**Limpeza e uniformização dos dados**

Exclusão de valores nulos e exclusão de serviços que não fazem parte do escopo "Uber" e "99" além da conversão de colunas de datas

```python

# Filtrar apenas Uber e 99
df_rideestimative = df_rideestimative[df_rideestimative['productid'].str.contains('UberX|99|pop99|top99|poupa99|Taxi|Bag|Black|Flash|Comfort|Uber', na=False)]
```

```python
# Excluir valores nulos em colunas importantes diretamente no df
df_rideaddress.dropna(subset=['lat', 'lng', 'number'], inplace=True)

# Conversão de colunas de datas para o date time e tratativa caso tenha valores inválidos
df_ride['schedule'] = pd.to_datetime(df_ride['schedule'], errors='coerce')
df_ride['create'] = pd.to_datetime(df_ride['create'], errors='coerce')
df_ride['updated'] = pd.to_datetime(df_ride['updated'], errors='coerce')
```

**Derivação de dados**

Nessa etapa derivamos o ano e o mês da corrida, bem como o dia da semana (útil para os dashs e treinamento de modelos) bem como a padronização de data e hora (exclusão de milisegundos)

```python
# Derivar ano e mês da corrida
df_ride['ano'] = df_ride['create'].dt.year
df_ride['mes'] = df_ride['create'].dt.month

# Derivar dia da semana:
# Foi criado um dicionário de dia da semana para utilização do pandas (pandas padroniza de 0 a 6) com o método dt.weekday.map()
dias_semana = {
    0: 'segunda-feira', 1: 'terça-feira', 2: 'quarta-feira',
    3: 'quinta-feira', 4: 'sexta-feira', 5: 'sábado', 6: 'domingo'
}
df_ride['dia_semana'] = df_ride['create'].dt.weekday.map(dias_semana)

# Data e horário formatados sem milissegundos
df_ride['data_hora_formatada'] = df_ride['create'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Somente data
df_ride['data'] = df_ride['create'].dt.date

# Tipo de endereço (origem/destino)
df_rideaddress['tipo_endereco'] = df_rideaddress['rideaddresstypeid'].map({1: 'origem', 2: 'destino'})

# Visualizar
display(df_ride.head(30))
```

**Formatação e integração**

Essa última etapa formatamos e integramos os dados para exibição em um único dataframe, como mencionado anteriormente tivemos o problema dos rideid se repetindo com as mesmas linhas (no qual tratamos na primeira linha de código) também foram renomeadas grande parte das colunas e integradas em uma única linha

```python
# Garantir que df_ride tenha rideid único, pois enfentamos um problema no qual os rideid se repetiam juntamente com suas linhas relacionadas
df_ride = df_ride.drop_duplicates(subset='rideid', keep='first')
```

```python
# Preparar e renomear a origem já que irá compor a mesma linha no df
df_origem = df_rideaddress[df_rideaddress['rideaddresstypeid'] == 1].copy() #para atuar com uma cópia e não uma visão
# Dicionário usado dentro do método rename para renomear
df_origem = df_origem.rename(columns={
    'lat': 'lat_origem',
    'lng': 'lng_origem',
    'number': 'number_origem',
    'rideaddresstypeid': 'rideaddresstypeid_origem'
}).drop_duplicates(subset='rideid', keep='first')

# Preparar e renomear o destino já que irá compor a mesma linha no df
df_destino = df_rideaddress[df_rideaddress['rideaddresstypeid'] == 2].copy()
df_destino = df_destino.rename(columns={
    'lat': 'lat_destino',
    'lng': 'lng_destino',
    'number': 'number_destino',
    'rideaddresstypeid': 'rideaddresstypeid_destino'
}).drop_duplicates(subset='rideid', keep='first')

```

```python
# Merge origem e destino com base no rideid usando inner para manter apenas os rideid correspondente em cada df
df_enderecos = pd.merge(df_origem, df_destino, on='rideid', how='inner')

# Merge com ride
df_base = pd.merge(df_ride, df_enderecos, on='rideid', how='inner')

# Merge com rideestimative
df_ride_full = pd.merge(df_base, df_rideestimative, on='rideid', how='inner')


# Ordenar as linhas com base no rideid e no price usando o inplace para modificar o df "original"
df_ride_full.sort_values(by=['rideid', 'price'], inplace=True)

```

```python
# Visualização (foi utilizado o max_columns pois algumas colunas não apareciam devido o tamanho dos dados)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
display(df_ride_full.head(60))
```
